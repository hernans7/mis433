```python
# Resolve this problem of "Gross" and convert its data type to float
# Hint: You may find this webpage useful:
# https://stackoverflow.com/questions/28986489/how-to-replace-text-in-a-column-of-a-pandas-dataframe

# Remove commas from the 'Gross' column and convert to numeric, coercing errors to NaN
imdb['Gross'] = imdb['Gross'].str.replace(',', '', regex=False).astype(float)

# Verify the change
display(imdb.info())
```

```python
# Add a new column "Runtime_min" by removing the substring " min" in "Runtime"
# Set its data type as int
# Hint: https://stackoverflow.stackoverflow.com/questions/36505847/substring-of-an-entire-column-in-pandas-dataframe

imdb['Runtime_min'] = imdb['Runtime'].str.replace(' min', '', regex=False).astype(int)

# Verify the change
display(imdb[['Runtime', 'Runtime_min']].head())
```

## Part 2: Data Summarization

```python
total_gross = imdb['Gross'].sum()
display(f"Total Gross of all top 1000 movies: ${total_gross:,.2f}")
```

```python
average_votes = imdb['No_of_Votes'].mean()
display(f"Average number of votes for all movies: {average_votes:,.2f}")
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
decade_counts = imdb['Decade'].value_counts().sort_values(ascending=False)
display(decade_counts)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
top_directors = imdb['Director'].value_counts().head(10)
display(top_directors)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
director_summary = imdb.groupby('Director').agg(
    num_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean'),
    total_gross=('Gross', 'sum')
).sort_values(by='total_gross', ascending=False)
display(director_summary.head(10))
```

```python
director_decade_summary = imdb.groupby(['Decade', 'Director']).agg(
    num_movies=('Series_Title', 'count'),
    average_imdb_rating=('IMDB_Rating', 'mean')
).sort_values(by='num_movies', ascending=False)
display(director_decade_summary.head())
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking

# Filter movies released since 2010
movies_since_2010 = imdb[imdb['Released_Year'] >= 2010]

# Group by Released_Year and find the top 3 highest rated movies for each year
top_movies_by_year = movies_since_2010.groupby('Released_Year').apply(
    lambda x: x.nlargest(3, 'IMDB_Rating', keep='first')
).reset_index(drop=True)

# Add a 'Ranking' column within each year
top_movies_by_year['Ranking'] = top_movies_by_year.groupby('Released_Year')['IMDB_Rating'].rank(ascending=False).astype(int)

# Select and sort the desired columns
result = top_movies_by_year[['Released_Year', 'Ranking', 'Series_Title', 'IMDB_Rating']]
result = result.sort_values(by=['Released_Year', 'Ranking'], ascending=[False, True])

display(result)
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.

# Group by director and decade, then count the number of movies
director_decade_counts = imdb.groupby(['Director', 'Decade']).agg(
    num_movies=('Series_Title', 'count')
).reset_index()

# Count the number of unique decades each director has movies in
decades_per_director = director_decade_counts.groupby('Director')['Decade'].nunique()

# Filter for directors who have movies in at least 5 different decades
directors_five_decades = decades_per_director[decades_per_director >= 5].index

# Filter the original director_decade_counts to include only these directors
result = director_decade_counts[director_decade_counts['Director'].isin(directors_five_decades)]

# Sort the results for better readability
result = result.sort_values(by=['Director', 'Decade'])

display(result)
```

```python
# Question 1: What is the average runtime (in minutes) for movies in each genre?
# Your result should include: Genre and Average Runtime.

average_runtime_by_genre = imdb.groupby('Genre')['Runtime_min'].mean().reset_index()
display(average_runtime_by_genre.sort_values(by='Runtime_min', ascending=False))
```

```python
# Question 2: Is there a correlation between IMDB_Rating and Gross revenue?
# Hint: Use the .corr() method after dropping NaN values.

correlation = imdb['IMDB_Rating'].corr(imdb['Gross'])
display(f"The correlation between IMDB_Rating and Gross revenue is: {correlation:.2f}")
```
