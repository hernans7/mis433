## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
#convert
flights['dep_time_min'] = flights ['dep_time_min']//100*60 + flights['dep_time']%100

flights['sched_dep_time_min'] = flights ['sched_dep_time_min']//100*60 + flights['sched_dep_time']%100

flights[['dep_time', 'sched_dep_time', 'dep_time_min', 'sched_dep_time_min']]
```

```python
def convert_hhmm_to_minutes(time_val):
    if pd.isna(time_val):
        return pd.NA
    time_str = str(int(time_val)).zfill(4)
    hours = int(time_str[:-2])
    minutes = int(time_str[-2:])
    return hours * 60 + minutes

# Convert dep_time and sched_dep_time to minutes since midnight
flights['dep_time_minutes'] = flights['dep_time'].apply(convert_hhmm_to_minutes)
flights['sched_dep_time_minutes'] = flights['sched_dep_time'].apply(convert_hhmm_to_minutes)
```

```python
# Convert arr_time to minutes since midnight
flights['arr_time_minutes'] = flights['arr_time'].apply(convert_hhmm_to_minutes)

# Create a new column of arr_time - dep_time in minutes
# Handle cases where arrival time might be on the next day (e.g., dep_time_minutes > arr_time_minutes)
# For simplicity, let's assume direct subtraction for now and acknowledge potential wrap-around issues
flights['flight_duration_calculated'] = flights['arr_time_minutes'] - flights['dep_time_minutes']

# Compare this column with air_time.
print("Comparing calculated flight duration with air_time:")
display(flights[['dep_time', 'arr_time', 'dep_time_minutes', 'arr_time_minutes', 'flight_duration_calculated', 'air_time']].head())
```

```python
# Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?


# Try creating a column to calculate dep_delay from dep_time and sched_dep_time (and/or other columns if necessary).
flights['calculated_dep_delay_minutes'] = flights.apply(
    lambda row: row['dep_time_minutes'] - row['sched_dep_time_minutes'] + (1440 if row['dep_time_minutes'] < row['sched_dep_time_minutes'] else 0)
    if pd.notna(row['dep_time_minutes']) and pd.notna(row['sched_dep_time_minutes']) else pd.NA, axis=1
)

# Test your results.
print("Comparing original dep_delay with calculated dep_delay_minutes:")
display(flights[['dep_time', 'sched_dep_time', 'dep_delay', 'dep_time_minutes', 'sched_dep_time_minutes', 'calculated_dep_delay_minutes']].head(10))


print("\nFlights where original dep_delay significantly differs from calculated_dep_delay_minutes (absolute difference > 5 minutes):")
display(flights[abs(flights['dep_delay'] - flights['calculated_dep_delay_minutes']) > 5].head())
```

```python
# Sort all AA flights to find the top 10 fastest (highest speed) flights.
# Display the following: year,month,day,carrier,flight,orig,dest,distance,air_time,speed (miles per hour)

# Filter for AA flights
aa_flights = flights[flights['carrier'] == 'AA'].copy()

# Calculate speed in miles per hour
# Ensure air_time is not zero to avoid division by zero errors
aa_flights['speed'] = aa_flights['distance'] / (aa_flights['air_time'] / 60)

# Sort by speed in descending order and get the top 10
fastest_aa_flights = aa_flights.sort_values(by='speed', ascending=False).head(10)

# Display the required columns
display(fastest_aa_flights[['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'distance', 'air_time', 'speed']])
```
